````julia
# Mutating vs non-mutating functions
# Mutatating functions return an altered version of the input: use !
# Non-mutating functions do NOT alter the input
# Example, sort vs sort! 
# push! and pop! were also mutating functions

v= [3,5,2]
println(sort(v))
v
````

````julia
println(sort!(v)) # notice that this actually changed v! Otherwise, we'd have to do v = sort(v)
v
````

````julia
# Broadcasting: use a ". to pass the operation to each element of the data struct
# Place between function and list of input
A = [i+j for i in 1:5,j in 1:5]
f = a -> a^2
````

````julia
# Not broadcast (returns the square of matrix A)
f(A)
````

````julia
# Broadcast (should return square of each element, not the square of matrix A)
f.(A)
````

````julia
# Notice that the square now DOES work for vectors, using broadcasting! No longer ambiguous
v = [1,2,3]
f.(v)
````

# Plotting
 Let's plot some polynomials; for example, $x^2$ (notice that we can use Tex!)

````julia
# you can use pyplot or Latex, or our nice Plots.jl which I just loaded above
# plots lets you change backends. The same commands will work for multiple backends
# gr() and plotlyjs() are both popular backends

x = -3:0.1:3 # a series of values between -3 and 3 in steps of 0.1
fff(x) = x^2
y= fff.(x) # broadcasting!
````

````julia
gr() # backend
````

````julia
plot(x,y,label="line")
scatter!(x,y,label="points") # adding the bang because we are mutating an existing plot. NO BANG = NEW PLOT
````

````julia
plotly() # a new backend! (for some reason plotlyjs wasn't there??)
````

````julia
# Using plotly we can now run the cursor over the data in the graph to get more info about it!
plot(x,y,label = "line")
scatter!(x,y,label="points")
````
