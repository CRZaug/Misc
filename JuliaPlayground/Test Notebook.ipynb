```julia
using Plots
```

```julia
function main(N::Any,L::Any) #Any: Float, array, etc. We don't want type inference 
    Δx = L/N
    x = collect(0.0:Δx:L-Δx) # iterator object; this won't allocate memory
    #x = Array{Float64}(undef, N) # Typing declaration; undef says whatever piece of memory the OS gives us, that is what we will use
    y = cos.(x)
    plot(x,y)
end
```

```julia
myPlot = main(1024, 2π)
```
